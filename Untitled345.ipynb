```python
import os
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# ============================================================
# PATHS
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System\archive\seattle-weather.csv"
BASE_DIR  = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System"
os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD & PREPROCESS DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

le = LabelEncoder()
df["weather"] = le.fit_transform(df["weather"])

X = df.drop(columns=["date", "weather"])
y = df["weather"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# FITNESS FUNCTION (MINIMIZE RMSE)
# ============================================================

def fitness(params):
    n_estimators = int(params[0])
    max_depth = int(params[1])

    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return np.sqrt(mean_squared_error(y_test, preds))

# ============================================================
# PSO OPTIMIZATION
# ============================================================

def PSO(particles=10, iterations=10):
    swarm = [
        [random.randint(50, 300), random.randint(3, 15)]
        for _ in range(particles)
    ]
    velocity = [[0, 0] for _ in range(particles)]
    pbest = swarm[:]
    gbest = min(swarm, key=fitness)

    for _ in range(iterations):
        for i in range(particles):
            for d in range(2):
                velocity[i][d] = (
                    0.5 * velocity[i][d]
                    + random.random() * (pbest[i][d] - swarm[i][d])
                    + random.random() * (gbest[d] - swarm[i][d])
                )
                swarm[i][d] += velocity[i][d]

                if d == 0:
                    swarm[i][d] = int(np.clip(swarm[i][d], 50, 300))
                else:
                    swarm[i][d] = int(np.clip(swarm[i][d], 3, 15))

            if fitness(swarm[i]) < fitness(pbest[i]):
                pbest[i] = swarm[i]

        gbest = min(pbest, key=fitness)

    return gbest

# ============================================================
# CSA OPTIMIZATION (LOCAL REFINEMENT)
# ============================================================

def CSA(initial, nests=10, iterations=10):
    population = [initial[:] for _ in range(nests)]

    for _ in range(iterations):
        for i in range(nests):
            step = [random.randint(-10, 10), random.randint(-2, 2)]
            candidate = [
                int(np.clip(population[i][0] + step[0], 50, 300)),
                int(np.clip(population[i][1] + step[1], 3, 15))
            ]
            if fitness(candidate) < fitness(population[i]):
                population[i] = candidate

    return min(population, key=fitness)

# ============================================================
# HYBRID PSO â†’ CSA EXECUTION
# ============================================================

pso_best = PSO()
best_params = CSA(pso_best)

print("âœ… Optimized Parameters (PSA):", best_params)

# ============================================================
# FINAL MODEL
# ============================================================

final_model = RandomForestRegressor(
    n_estimators=int(best_params[0]),
    max_depth=int(best_params[1]),
    random_state=42
)

final_model.fit(X_train, y_train)
y_pred = final_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

# ============================================================
# SAVE RESULTS CSV
# ============================================================

pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
}).to_csv(
    os.path.join(BASE_DIR, "psa_tree_results.csv"),
    index=False
)

# ============================================================
# SAVE PREDICTION JSON
# ============================================================

with open(
    os.path.join(BASE_DIR, "psa_tree_predictions.json"),
    "w"
) as f:
    json.dump({
        "rmse": float(rmse),
        "r2_score": float(r2),
        "best_parameters": {
            "n_estimators": int(best_params[0]),
            "max_depth": int(best_params[1])
        },
        "optimizer": "PSO + CSA"
    }, f, indent=4)

# ============================================================
# VISUALIZATIONS (SHOW + SAVE)
# ============================================================

plt.figure()
plt.bar(["Hybrid PSO+CSA"], [r2])
plt.title("PSA Model Accuracy (RÂ²)")
plt.savefig(os.path.join(BASE_DIR, "psa_tree_accuracy.png"))
plt.show()

plt.figure()
plt.bar(["Actual Mean", "Predicted Mean"],
        [np.mean(y_test), np.mean(y_pred)])
plt.title("PSA Actual vs Predicted Mean")
plt.savefig(os.path.join(BASE_DIR, "psa_tree_comparison.png"))
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=["date"]).corr(), annot=True, cmap="coolwarm")
plt.title("PSA Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR, "psa_tree_heatmap.png"))
plt.show()

plt.figure()
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("PSA Actual vs Predicted")
plt.savefig(os.path.join(BASE_DIR, "psa_tree_actual_vs_predicted.png"))
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(y_test.values[:100], label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.legend()
plt.title("PSA Prediction Trend")
plt.savefig(os.path.join(BASE_DIR, "psa_tree_prediction_trend.png"))
plt.show()

# ============================================================
# DONE
# ============================================================

print("ðŸ”¥ HYBRID PSO + CSA PIPELINE COMPLETE")
print("ðŸ“Š RMSE:", rmse)
print("ðŸ“ˆ RÂ²:", r2)
print("ðŸ“ All outputs saved with 'psa_' prefix")
```
