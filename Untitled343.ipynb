```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# ============================================================
# PATHS
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System\archive\seattle-weather.csv"
BASE_DIR  = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System"
os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD & PREPROCESS DATA
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

le = LabelEncoder()
df["weather"] = le.fit_transform(df["weather"])

X = df.drop(columns=["date", "weather"])
y = df["weather"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# TRAIN MODEL
# ============================================================

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    random_state=42
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# ============================================================
# METRICS
# ============================================================

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

# ============================================================
# SAVE RESULT CSV
# ============================================================

results_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

results_csv = os.path.join(BASE_DIR, "tree_results.csv")
results_df.to_csv(results_csv, index=False)

# ============================================================
# SAVE PREDICTION JSON
# ============================================================

pred_json = {
    "rmse": float(rmse),
    "r2_score": float(r2),
    "predictions": y_pred.tolist()
}

json_path = os.path.join(BASE_DIR, "tree_predictions.json")
with open(json_path, "w") as f:
    json.dump(pred_json, f, indent=4)

# ============================================================
# ACCURACY GRAPH (R2)
# ============================================================

plt.figure()
plt.bar(["Random Forest"], [r2])
plt.title("Model Accuracy (R Score)")
plt.ylabel("R")
plt.ylim(0, 1)
plt.savefig(os.path.join(BASE_DIR, "accuracy_comparison.png"))
plt.show()

# ============================================================
# COMPARISON GRAPH (Actual vs Predicted Mean)
# ============================================================

plt.figure()
plt.bar(
    ["Actual Mean", "Predicted Mean"],
    [np.mean(y_test), np.mean(y_pred)]
)
plt.title("Actual vs Predicted Mean Comparison")
plt.savefig(os.path.join(BASE_DIR, "model_comparison.png"))
plt.show()

# ============================================================
# HEATMAP (Feature Correlation)
# ============================================================

plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=["date"]).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR, "correlation_heatmap.png"))
plt.show()

# ============================================================
# ACTUAL vs PREDICTED SCATTER
# ============================================================

plt.figure()
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Scatter Plot")
plt.savefig(os.path.join(BASE_DIR, "actual_vs_predicted.png"))
plt.show()

# ============================================================
# PREDICTION TREND GRAPH
# ============================================================

plt.figure(figsize=(10, 4))
plt.plot(y_test.values[:100], label="Actual", marker="o")
plt.plot(y_pred[:100], label="Predicted", marker="x")
plt.legend()
plt.title("Prediction Trend (First 100 Samples)")
plt.savefig(os.path.join(BASE_DIR, "prediction_trend.png"))
plt.show()

# ============================================================
# DONE
# ============================================================

print(" All graphs generated, displayed, and saved")
print(" Results saved in:", BASE_DIR)
print(" RMSE:", rmse)
print(" R2 Score:", r2)
```
