```python
import os
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# ============================================================
# PATHS
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System\archive\seattle-weather.csv"
BASE_DIR  = r"C:\Users\NXTWAVE\Downloads\Urban Tree Canopy Optimization System"
os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD & PREPROCESS
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

le = LabelEncoder()
df["weather"] = le.fit_transform(df["weather"])

X = df.drop(columns=["date", "weather"])
y = df["weather"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(params):
    n_estimators = int(params[0])
    max_depth = int(params[1])

    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=42
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    return rmse

# ============================================================
# AIS OPTIMIZATION
# ============================================================

def AIS(pop_size=10, generations=10):
    population = [
        [random.randint(50, 300), random.randint(3, 15)]
        for _ in range(pop_size)
    ]

    for _ in range(generations):
        scores = [(fitness(p), p) for p in population]
        scores.sort(key=lambda x: x[0])
        elites = [p for _, p in scores[:5]]

        clones = []
        for p in elites:
            for _ in range(2):
                clone = [
                    max(50, min(300, p[0] + random.randint(-20, 20))),
                    max(3, min(15, p[1] + random.randint(-2, 2)))
                ]
                clones.append(clone)

        population = elites + clones

    best = min(population, key=fitness)
    return best

# ============================================================
# CSA OPTIMIZATION
# ============================================================

def CSA(init_solution, nests=10, iterations=10):
    population = [init_solution[:] for _ in range(nests)]

    for _ in range(iterations):
        for i in range(nests):
            step = [random.randint(-10, 10), random.randint(-1, 1)]
            candidate = [
                max(50, min(300, population[i][0] + step[0])),
                max(3, min(15, population[i][1] + step[1]))
            ]

            if fitness(candidate) < fitness(population[i]):
                population[i] = candidate

    best = min(population, key=fitness)
    return best

# ============================================================
# HYBRID EXECUTION
# ============================================================

ais_best = AIS()
best_params = CSA(ais_best)

print("âœ… Optimized Parameters:", best_params)

# ============================================================
# FINAL MODEL
# ============================================================

final_model = RandomForestRegressor(
    n_estimators=int(best_params[0]),
    max_depth=int(best_params[1]),
    random_state=42
)

final_model.fit(X_train, y_train)
y_pred = final_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# ============================================================
# SAVE RESULTS CSV
# ============================================================

results_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

results_df.to_csv(
    os.path.join(BASE_DIR, "hybrid_tree_results.csv"),
    index=False
)

# ============================================================
# SAVE PREDICTIONS JSON
# ============================================================

with open(
    os.path.join(BASE_DIR, "hybrid_tree_predictions.json"),
    "w"
) as f:
    json.dump({
        "rmse": float(rmse),
        "r2_score": float(r2),
        "best_parameters": {
            "n_estimators": int(best_params[0]),
            "max_depth": int(best_params[1])
        }
    }, f, indent=4)

# ============================================================
# VISUALIZATIONS (SHOW + SAVE)
# ============================================================

plt.figure()
plt.bar(["Hybrid AIS+CSA"], [r2])
plt.title("Hybrid Model Accuracy (RÂ²)")
plt.savefig(os.path.join(BASE_DIR, "hybrid_tree_accuracy.png"))
plt.show()

plt.figure()
plt.bar(["Actual Mean", "Predicted Mean"], [np.mean(y_test), np.mean(y_pred)])
plt.title("Hybrid Actual vs Predicted Mean")
plt.savefig(os.path.join(BASE_DIR, "hybrid_tree_comparison.png"))
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=["date"]).corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR, "hybrid_tree_heatmap.png"))
plt.show()

plt.figure()
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Hybrid Actual vs Predicted")
plt.savefig(os.path.join(BASE_DIR, "hybrid_tree_actual_vs_predicted.png"))
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(y_test.values[:100], label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.legend()
plt.title("Hybrid Prediction Trend")
plt.savefig(os.path.join(BASE_DIR, "hybrid_tree_prediction_trend.png"))
plt.show()

# ============================================================
# DONE
# ============================================================

print("ðŸ”¥ HYBRID AIS + CSA PIPELINE COMPLETE")
print("ðŸ“Š RMSE:", rmse)
print("ðŸ“ˆ RÂ²:", r2)
print("ðŸ“ Saved all results with 'hybrid_tree_' prefix")
```
